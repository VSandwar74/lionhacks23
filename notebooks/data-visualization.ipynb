```python
def tweets_dist(data):
    fig = go.Figure()

    fig.add_trace(go.Histogram(
      x=data['BLOCK_TIMESTAMP'],
      nbinsx=30
    ))

    fig.update_layout(
      title_text='Distribution of tweets over time',
      xaxis_title_text='Time',
      yaxis_title_text='Frequency',
      bargap=0.1,
      bargroupgap=0.2)
    
    return fig

#sample use
tweets_dist(data).show()
```

```python
grouping_freq='20min'
```

```python
data['BLOCK_TIMESTAMP'] = pd.to_datetime(data['BLOCK_TIMESTAMP'])
grouped_negative = data[data['sentiment'] == 'negative'].groupby(pd.Grouper(key='BLOCK_TIMESTAMP',freq=grouping_freq)).count().head()
grouped_neutral = data[data['sentiment'] == 'neutral'].groupby(pd.Grouper(key='BLOCK_TIMESTAMP',freq=grouping_freq)).count().head()
grouped_positive = data[data['sentiment'] == 'positive'].groupby(pd.Grouper(key='BLOCK_TIMESTAMP',freq=grouping_freq)).count().head()
grouped_total = data.groupby(pd.Grouper(key='BLOCK_TIMESTAMP',freq=grouping_freq)).count().head()
final_data = pd.DataFrame(data={
    'negative': grouped_negative['sentiment'],
    'neutral' : grouped_neutral['sentiment'],
    'positive': grouped_positive['sentiment'],
},index=grouped_total.index)
```

```python
final_data.reset_index(inplace=True)
```

```python
fig = go.Figure()
fig.add_trace(go.Scatter(
    x=final_data['BLOCK_TIMESTAMP'],
    y=final_data['neutral'],
    name="Neutrals",
    opacity=0.8,
    mode='lines',
    line=dict(width=0.5, color='rgb(131, 90, 241)'),
    stackgroup='one' 
))
fig.add_trace(go.Scatter(
    x=final_data['BLOCK_TIMESTAMP'],
    y=final_data['negative'],
    name="Negatives",
    opacity=0.8,
    mode='lines',
    line=dict(width=0.5, color='rgb(255, 50, 50)'),
    stackgroup='two' 
))
fig.add_trace(go.Scatter(
    x=final_data['BLOCK_TIMESTAMP'],
    y=final_data['positive'],
    name="Positives",
    opacity=0.8,
    mode='lines',
    line=dict(width=0.5, color='rgb(184, 247, 212)'),
    stackgroup='three' 
))
fig.update_layout(
    title_text='Distribution of sentiments over time',
    xaxis_title_text='Time',
    yaxis_title_text='Frequency'
)
fig.update_xaxes(
    rangeslider_visible=True
)
fig.show()
```

```python
def sentiment_dist(data,grouping_freq='10min'):
  grouped_negative = data[data['sentiment'] == 'negative'].groupby(pd.Grouper(key='BLOCK_TIMESTAMP',freq=grouping_freq)).count()
  grouped_neutral = data[data['sentiment'] == 'neutral'].groupby(pd.Grouper(key='BLOCK_TIMESTAMP',freq=grouping_freq)).count()
  grouped_positive = data[data['sentiment'] == 'positive'].groupby(pd.Grouper(key='BLOCK_TIMESTAMP',freq=grouping_freq)).count()
  grouped_total = data.groupby(pd.Grouper(key='BLOCK_TIMESTAMP',freq=grouping_freq)).count()
  final_data = pd.DataFrame(data={
      'negative': grouped_negative['sentiment'],
      'neutral' : grouped_neutral['sentiment'],
      'positive': grouped_positive['sentiment'],
  },index=grouped_total.index)
  final_data.reset_index(inplace=True)
  fig = go.Figure()
  fig.add_trace(go.Scatter(
      x=final_data['BLOCK_TIMESTAMP'],
      y=final_data['neutral'],
      name="Neutrals",
      opacity=0.8,
      mode='lines',
      line=dict(width=0.5, color='rgb(131, 90, 241)'),
      stackgroup='one' 
  ))
  fig.add_trace(go.Scatter(
      x=final_data['BLOCK_TIMESTAMP'],
      y=final_data['negative'],
      name="Negatives",
      opacity=0.8,
      mode='lines',
      line=dict(width=0.5, color='rgb(255, 50, 50)'),
      stackgroup='two' 
  ))
  fig.add_trace(go.Scatter(
      x=final_data['BLOCK_TIMESTAMP'],
      y=final_data['positive'],
      name="Positives",
      opacity=0.8,
      mode='lines',
      line=dict(width=0.5, color='rgb(184, 247, 212)'),
      stackgroup='three' 
  ))
  fig.update_layout(
      title_text='Distribution of sentiments over time',
      xaxis_title_text='Time',
      yaxis_title_text='Frequency'
  )
  fig.update_xaxes(
      rangeslider_visible=True
  )
  return fig
```
